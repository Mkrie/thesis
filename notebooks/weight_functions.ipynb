### Послойные воздушные массы для различных углов направления визирования

```python
import matplotlib.pyplot as plt
import csv
from pathlib import Path
```

```python
plt.figure(figsize=(15, 10))
for txt_path in Path(Path.cwd().parent, 
                     "data", "dataset_1_csv").glob("*.csv"):
    with open(txt_path, "r") as csv_file:
        reader = csv.reader(csv_file)
        tup_coords: tuple[tuple[str]] = tuple(reader)
        plt.grid()
        plt.ylim(0, 32)
        plt.xlim(0, 1000)
        plt.xlabel("height[m]")
        plt.ylabel("$m$")
        #plt.title(txt_path.name.replace(".csv", "").replace("_", " "))
        label=txt_path.name.replace(".csv", "").replace("_", " ").split()[1:]
        if len(label)==1:
            label=f"{label[0]}°"
        else:
            label=f"{label[0]},{label[1]}°"
        plt.plot(tuple(float(x[0])*1000 for x in tup_coords), 
                 tuple(float(x[1]) for x in tup_coords),
                 linewidth=3,
                 label=label)
        #plt.xscale('log')
        #print(txt_path.name)

plt.legend()
plt.show()
```
